```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV
df = pd.read_csv(r"C:\Users\exam\Downloads\supermarket_sales - Sheet1.csv")

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group by Date and calculate total sales per day
daily_sales = df.groupby('Date')['Total'].sum().sort_index()

# Apply 7-day moving average
window_size = 7
moving_avg = daily_sales.rolling(window=window_size).mean()

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(daily_sales, label='Original Daily Sales')
plt.plot(moving_avg, label=f'{window_size}-Day Moving Average', linewidth=2)
plt.title('Daily Sales with Moving Average Smoothing')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
